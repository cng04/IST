```python
import pandas as pd
```

```python
player_box_score_stats = pd.read_excel("data/usports_stats_data.xlsx")
player_box_score_stats
```

```python
manually_scraped_player = pd.read_excel("data/players_missing_usports_stats.xlsx")
manually_scraped_player
```

```python
df = pd.concat([player_box_score_stats, manually_scraped_player], ignore_index=True)
df
```

```python
usports_to_pro_summary = pd.read_excel("data/usports_players_to_pro_summary.xlsx")

# Merge with usports_to_pro_summary
df = df.merge(
    usports_to_pro_summary[["Player", "Position", "Hometown", "U Sports Team"]],
    on="Player",
    how="left"
)

cols = (
    ["Player", "Position", "Hometown", "U Sports Team"]
    + [c for c in df.columns if c not in ["Player", "Position", "Hometown", "U Sports Team"]]
)

df = df[cols]
df
```

```python
# Drop Unnecessary Columns
df = df.drop(columns=["Mins", "Rebounds_1", "Points_1"])
df
```

```python
# Split Position into Position and Height
split_cols = df["Position"].str.split(" ", n=2, expand=True)
pos_map = {"Guard": "G", "Forward": "F", "Centre": "C", "Wing": "G/F", "Guard/Wing": "G/F", "Ailier": "SF", "Arriere": "SG", "Meneur": "PG", "Centre/Forward": "C/F", "Garde": "G", "Guard-Forward": "G/F", "Guard/Forward": "G/F", "Post": "F/C", "Point Guard": "PG"}

def map_position(row):
    # try two-word combo
    if pd.notna(row[1]):
        two_word = f"{row[0]} {row[1]}"
        if two_word in pos_map:
            return pos_map[two_word], row[2]  # use 3rd column as height
    # otherwise, use single-word
    return pos_map.get(row[0], row[0]), row[1]

df[["Position", "Height"]] = split_cols.apply(map_position, axis=1, result_type="expand")
cols = list(df.columns)
pos_idx = cols.index("Position")
cols.remove("Height")
cols.insert(pos_idx + 1, "Height")

# reorder dataframe
df = df[cols]
df
```

```python
# Split GP-GS column
split_cols = df["GP-GS"].str.split("-", n=1, expand=True)
gp_gs_index = df.columns.get_loc("GP-GS")
df.insert(gp_gs_index + 1, "GP", pd.to_numeric(split_cols[0], errors="coerce").astype("Int64"))
df.insert(gp_gs_index + 2, "GS", pd.to_numeric(split_cols[1], errors="coerce").astype("Int64"))
df = df.drop(columns=["GP-GS"])
df
```

```python
# Split 3PT column
split_cols = df["3 Pt_1"].str.split("-", n=1, expand=True)
index = df.columns.get_loc("3 Pt_1")
df.insert(index + 1, "3PM", pd.to_numeric(split_cols[0], errors="coerce").astype("Int64"))
df.insert(index + 2, "3PA", pd.to_numeric(split_cols[1], errors="coerce").astype("Int64"))
df = df.drop(columns=["3 Pt_1"])

# Rename 3PT% Column
df = df.rename(columns={"3 Pt_2": "3P%"})
df["3P%"] = df["3P%"].astype(float)
df
```

```python
# Split Field Goals column
split_cols = df["Field Goals_1"].str.split("-", n=1, expand=True)
index = df.columns.get_loc("Field Goals_1")
df.insert(index + 1, "FGM", pd.to_numeric(split_cols[0], errors="coerce").astype("Int64"))
df.insert(index + 2, "FGA", pd.to_numeric(split_cols[1], errors="coerce").astype("Int64"))
df = df.drop(columns=["Field Goals_1"])

# Rename FG% Column
df = df.rename(columns={"Field Goals_2": "FG%"})
df["FG%"] = df["FG%"].astype(float)
df[["Player", "FGM", "FGA", "FG%"]].head(20)
```

```python
# Split Field Throw column
split_cols = df["Free Throws_1"].str.split("-", n=1, expand=True)
index = df.columns.get_loc("Free Throws_1")
df.insert(index + 1, "FTM", pd.to_numeric(split_cols[0], errors="coerce").astype("Int64"))
df.insert(index + 2, "FTA", pd.to_numeric(split_cols[1], errors="coerce").astype("Int64"))
df = df.drop(columns=["Free Throws_1"])

# Rename FT% Column
df = df.rename(columns={"Free Throws_2": "FT%"})
df["FT%"] = df["FT%"].astype(float)
df[["Player", "FTM", "FTA", "FT%"]].head(20)
```

```python
# Rename Columns
df = df.rename(columns={"Rebounds_2": "REB", "Points_2": "PPG", "AS": "AST", "TO": "TOV", "BL": "BLK", "ST": "STL"})
df
```

```python
# Calculating per game stats
df["PF"] = (df["PF"] / df["GP"]).round(1)
df["AST"] = (df["AST"] / df["GP"]).round(1)
df["TOV"] = (df["TOV"] / df["GP"]).round(1)
df["BLK"] = (df["BLK"] / df["GP"]).round(1)
df["STL"] = (df["STL"] / df["GP"]).round(1)
df
```

```python
df.to_excel("data/cleaned_usports_player_box_score_stats.xlsx", index=False, engine="openpyxl")
```

```python
# Load career data for merging with cleaned stats
career_df = pd.read_excel("data/existing_data/usports_career_data.xlsx")
career_df.head()
```

```python
def merge_career_with_stats(career_df, stats_df):
    """Merge career data with USports stats based on season and player."""
    if career_df.empty or stats_df.empty:
        print("Cannot merge: one or both dataframes are empty")
        return pd.DataFrame()
    
    # Merge on Player and Season columns
    try:
        # Ensure Season column exists in stats_df
        if "Season" not in stats_df.columns:
            print("Warning: Season column not found in stats data. Cannot merge.")
            return stats_df
        
        career_df["Team"] = career_df["Team"].str.replace("MBB-", "", regex=True)
            
        merged_df = pd.merge(stats_df, career_df[["Player", "Season", "Team"]], 
                            on=["Player", "Season"], 
                            how="left")
        print(f"Successfully merged career and stats data. Result has {len(merged_df)} rows.")
        return merged_df
        
    except Exception as e:
        print(f"Error merging career and stats data: {e}")
        return stats_df

# Merge career data with cleaned stats
print("Merging career data with cleaned stats...")
merged_df = merge_career_with_stats(career_df, df)
merged_df.head()
```

```python
def validate_player_counts(usports_stats_with_teams_df, pro_season_data_df):
    """Validate that the number of unique players in both datasets is equal."""
    if usports_stats_with_teams_df.empty or pro_season_data_df.empty:
        print("Warning: Cannot validate player counts - one or both datasets are empty")
        return False
    
    # Get unique player counts from both datasets
    usports_unique_players = usports_stats_with_teams_df['Player'].nunique()
    pro_unique_players = pro_season_data_df['Player'].nunique()
    
    print(f"Unique players in usports_stats_with_teams: {usports_unique_players}")
    print(f"Unique players in pro_season_data: {pro_unique_players}")
    
    if usports_unique_players == pro_unique_players:
        print("✓ VALIDATION PASSED: Player counts match!")
        return True
    else:
        print("✗ VALIDATION FAILED: Player counts do not match!")
        
        # Show which players are missing from each dataset
        usports_players = set(usports_stats_with_teams_df['Player'].unique())
        pro_players = set(pro_season_data_df['Player'].unique())
        
        missing_from_usports = pro_players - usports_players
        missing_from_pro = usports_players - pro_players
        
        if missing_from_usports:
            print(f"Players in pro_season_data but not in usports_stats_with_teams: {missing_from_usports}")
        if missing_from_pro:
            print(f"Players in usports_stats_with_teams but not in pro_season_data: {missing_from_pro}")
        
        return False

# Load pro season data for validation
try:
    pro_seasons_df = pd.read_excel("data/pro_season_data.xlsx")
    print("Validating player counts...")
    validate_player_counts(merged_df, pro_seasons_df)
except FileNotFoundError:
    print("Pro season data file not found. Skipping validation.")
except Exception as e:
    print(f"Error loading pro season data: {e}")
```

```python
# Rename columns as requested
merged_df = merged_df.rename(columns={
    "U Sports Team": "All USports Teams Played For",
    "Team": "USports Team"
})

# Reorder columns to move USports Team to the right of Hometown
cols = list(merged_df.columns)
hometown_idx = cols.index("Hometown")
usports_team_idx = cols.index("USports Team")

# Remove USports Team from its current position
cols.remove("USports Team")
# Insert it right after Hometown
cols.insert(hometown_idx + 1, "USports Team")

# Reorder the dataframe
merged_df = merged_df[cols]

# Save the final merged dataset
if not merged_df.empty:
    merged_df.to_excel("data/usports_stats_with_teams.xlsx", index=False, engine="openpyxl")
    print("Final merged dataset saved to data/usports_stats_with_teams.xlsx")
    print(f"Dataset contains {len(merged_df)} rows and {len(merged_df.columns)} columns")
    print(f"Unique players: {merged_df['Player'].nunique()}")
    print(f"Column order: {list(merged_df.columns[:8])}")  # Show first 8 columns to verify order
else:
    print("Merged dataframe is empty - not saving")
```

```python
# Create analysis of average stats by number of USports seasons played
def create_pro_transition_analysis(usports_df):
    """
    Create a dataframe showing average stats of players who went pro,
    grouped by the number of USports seasons they played.
    """
    
    # Filter out 'Total' rows - we only want individual seasons
    usports_seasons_only = usports_df[usports_df['Season'] != 'Total'].copy()
    
    # For each player, count how many seasons they played and get their average stats
    player_analysis = []
    
    for player in usports_seasons_only['Player'].unique():
        player_data = usports_seasons_only[usports_seasons_only['Player'] == player]
        num_seasons = len(player_data)
        
        # Calculate average stats across all seasons for this player
        numeric_cols = ['MPG', '3PM', '3PA', '3P%', 'FGM', 'FGA', 'FG%', 'FTM', 'FTA', 'FT%', 
                       'REB', 'PF', 'AST', 'TOV', 'BLK', 'STL', 'PPG']
        
        player_avg_stats = {}
        player_avg_stats['Player'] = player
        player_avg_stats['Num_USports_Seasons'] = num_seasons
        
        for col in numeric_cols:
            if col in player_data.columns:
                avg_val = player_data[col].mean()
                player_avg_stats[f'Avg_{col}'] = avg_val
        
        player_analysis.append(player_avg_stats)
    
    player_stats_df = pd.DataFrame(player_analysis)
    
    # Group by number of seasons and calculate overall averages
    season_groups = player_stats_df.groupby('Num_USports_Seasons')
    
    # Calculate averages for each group
    avg_stats_by_seasons = []
    
    for num_seasons, group in season_groups:
        stats_summary = {'Num_USports_Seasons': num_seasons, 'Player_Count': len(group)}
        
        for col in player_stats_df.columns:
            if col.startswith('Avg_'):
                stat_name = col.replace('Avg_', '')
                avg_value = group[col].mean()
                stats_summary[stat_name] = round(avg_value, 1)
        
        avg_stats_by_seasons.append(stats_summary)
    
    pro_transition_stats = pd.DataFrame(avg_stats_by_seasons)
    pro_transition_stats = pro_transition_stats.sort_values('Num_USports_Seasons')
    
    print(f"Analysis complete:")
    print(f"Total unique players: {len(player_stats_df)}")
    print(f"Season ranges: {pro_transition_stats['Num_USports_Seasons'].min()} to {pro_transition_stats['Num_USports_Seasons'].max()} seasons")
    
    return pro_transition_stats

# Run the analysis
pro_transition_stats_df = create_pro_transition_analysis(merged_df)
pro_transition_stats_df
```

```python
# Export the pro transition analysis to its own Excel file
pro_transition_stats_df.to_excel("data/pro_transition_stats_by_seasons.xlsx", index=False, engine="openpyxl")
print("Pro transition analysis exported to data/pro_transition_stats_by_seasons.xlsx")
print(f"Dataset contains {len(pro_transition_stats_df)} rows (different season counts)")
print("\nSample of the analysis:")
print(pro_transition_stats_df.head())
```
