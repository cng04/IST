```python
import pandas as pd
```

```python
df = pd.read_excel("data/current_usports_players.xlsx")
df
```

```python
# Drop unnecessary columns
df = df.drop(columns=["stats_Mins_0", "stats_Rebounds_0", "stats_Points_0"])
df
```

```python
def splitting_data(df: pd.DataFrame, games_column_name: str, three_pt_column_name: str, three_pt_pct_column_name: str, field_goals_column_name: str, field_goals_pct_column_name: str, field_throws_column_name: str, field_throws_pct_column_name: str) -> pd.DataFrame:
    split_cols = df[games_column_name].str.split("-", n=1, expand=True)
    gp_gs_index = df.columns.get_loc(games_column_name)
    df.insert(gp_gs_index + 1, "GP", pd.to_numeric(split_cols[0], errors="coerce").astype("Int64"))
    df.insert(gp_gs_index + 2, "GS", pd.to_numeric(split_cols[1], errors="coerce").astype("Int64"))
    df = df.drop(columns=[games_column_name])

    # Split 3PT column
    split_cols = df[three_pt_column_name].str.split("-", n=1, expand=True)
    index = df.columns.get_loc(three_pt_column_name)
    df.insert(index + 1, "3PM", pd.to_numeric(split_cols[0], errors="coerce").astype("Int64"))
    df.insert(index + 2, "3PA", pd.to_numeric(split_cols[1], errors="coerce").astype("Int64"))
    df = df.drop(columns=[three_pt_column_name])

    # Rename 3PT% Column
    df = df.rename(columns={three_pt_pct_column_name: "3P%"})
    df["3P%"] = df["3P%"].astype(float)

    # Split Field Goals column
    split_cols = df[field_goals_column_name].str.split("-", n=1, expand=True)
    index = df.columns.get_loc(field_goals_column_name)
    df.insert(index + 1, "FGM", pd.to_numeric(split_cols[0], errors="coerce").astype("Int64"))
    df.insert(index + 2, "FGA", pd.to_numeric(split_cols[1], errors="coerce").astype("Int64"))
    df = df.drop(columns=[field_goals_column_name])

    # Rename FG% Column
    df = df.rename(columns={field_goals_pct_column_name: "FG%"})
    df["FG%"] = df["FG%"].astype(float)

    # Split Field Throw column
    split_cols = df[field_throws_column_name].str.split("-", n=1, expand=True)
    index = df.columns.get_loc(field_throws_column_name)
    df.insert(index + 1, "FTM", pd.to_numeric(split_cols[0], errors="coerce").astype("Int64"))
    df.insert(index + 2, "FTA", pd.to_numeric(split_cols[1], errors="coerce").astype("Int64"))
    df = df.drop(columns=[field_throws_column_name])

    # Rename FT% Column
    df = df.rename(columns={field_throws_pct_column_name: "FT%"})
    df["FT%"] = df["FT%"].astype(float)

    # Remove MBB- from team names
    df["career_team"] = df["career_team"].str.replace("MBB-", "", regex=False)

    return df
```

```python
df = splitting_data(df, "stats_GP-GS_0", "stats_3 Pt_0", "stats_3 Pt_1", "stats_Field Goals_0", "stats_Field Goals_1", "stats_Free Throws_0", "stats_Free Throws_1")

df
```

```python
def rename_columns(df: pd.DataFrame, columns: list[str], new_columns: list[str]) -> pd.DataFrame:
    rename_dict = dict(zip(columns, new_columns))
    df = df.rename(columns=rename_dict)
    return df
```

```python
df = rename_columns(df, ["name", "career_season", "career_team", "career_elig", "stats_Season_0", "stats_MPG_0", "stats_Rebounds_1", "stats_PF_0", "stats_AS_0", "stats_TO_0", "stats_BL_0", "stats_ST_0", "stats_Points_1"],
                            ["Player", "Season(s)", "Team", "Eligibility", "Season", "MPG", "REB", "PF", "AST", "TOV", "BLK", "STL", "PTS"])
df
```

```python
def calculate_averages(df: pd.DataFrame, columns: list[str]) -> pd.DateOffset:
    for column in columns:
        df[column] = (df[column] / df["GP"]).round(1)

    return df
```

```python
df = calculate_averages(df, ["PF", "AST", "TOV", "BLK", "STL"])
df
```

```python
df.to_excel("data/processed_usports_players.xlsx", index=False, engine="openpyxl")
```
